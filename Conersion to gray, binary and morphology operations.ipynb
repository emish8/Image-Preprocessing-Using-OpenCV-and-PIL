# Image Processing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python
img = cv2.imread('images.png')
img = cv2.cvtColor(img,  cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(img)
```

# Black White Image Conversion

```python
# 1. Using opencv
bw_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(3,3))
plt.imshow(bw_img, cmap = 'gray')
```

```python
# 2. Averaging

r,g,b = img[:,:,0] , img[:,:,1], img[:,:,2]
avg_img = r/3 + g/3 + b/3
plt.figure(figsize=(3,3))
plt.imshow(avg_img, cmap = 'gray')
```

```python
r,g,b = img[:,:,0] , img[:,:,1], img[:,:,2]
wavg_img = (.299*r/3) + (.588*g/3) + (.114*b/3)
plt.figure(figsize=(3,3))
plt.imshow(wavg_img, cmap = 'gray')
```

```python
# 4. negative Image
neg_img = img - 255
plt.figure(figsize=(3,3))
plt.imshow(neg_img, cmap = 'gray')
```

```python
plt.figure(figsize=(12,5))
plt.subplot(141).set_title('opencv'), plt.imshow(bw_img, cmap = 'gray')
plt.subplot(142).set_title('avg_img'), plt.imshow(avg_img, cmap = 'gray')
plt.subplot(143).set_title('wieghteh avg'), plt.imshow(wavg_img, cmap = 'gray')
plt.subplot(144).set_title('negative '), plt.imshow(neg_img, cmap = 'gray')
```

# Morphology

```python
kernel = np.ones((5,5)).astype(int)
kernel
```

```python
# 1. erosion
ero_img = cv2.erode(bw_img, kernel = kernel, iterations = 1)
plt.figure(figsize=(3,3))
plt.imshow(ero_img, cmap = 'gray')
```

```python
# 2. Dilation
dil_img = cv2.dilate(bw_img, kernel = kernel, iterations=1)
plt.figure(figsize=(3,3))
plt.imshow(dil_img, cmap = 'gray')
```

```python
# 3. opening

op_img = cv2.morphologyEx(bw_img, cv2.MORPH_OPEN, kernel = kernel)
plt.figure(figsize=(3,3))
plt.imshow(op_img, cmap = 'gray')
```

```python
# 4. closing
cl_img = cv2.morphologyEx(bw_img, cv2.MORPH_CLOSE, kernel = kernel)
plt.figure(figsize=(3,3))
plt.imshow(cl_img, cmap = 'gray')
```

```python
# 5.  top-hat 
# top hat = input_img - open_img

th_img = cv2.morphologyEx(bw_img, cv2.MORPH_TOPHAT, kernel=kernel)
plt.figure(figsize=(3,3))
plt.imshow(th_img, cmap = 'gray')
```

```python
# balck hat
# black hat = close_img - original_img
bh_img = cv2.morphologyEx(bw_img,cv2.MORPH_BLACKHAT, kernel = kernel )
plt.figure(figsize=(3,3))
plt.imshow(bh_img, cmap = 'gray')
```

```python
# 6. Gradient 
gd_img = cv2.morphologyEx(bw_img, cv2.MORPH_GRADIENT, kernel = kernel)
plt.figure(figsize=(3,3))
plt.imshow(bh_img, cmap = 'gray')
```

```python
# plotting for comparison
plt.figure(figsize=(12,6))
    
plt.subplot(2,4,1).set_title('black and white'),plt.imshow(bw_img, cmap = 'gray')
plt.subplot(2,4,2).set_title('erosion'),plt.imshow(ero_img, cmap = 'gray')
plt.subplot(2,4,3).set_title('dilation'),plt.imshow(dil_img, cmap = 'gray')
plt.subplot(2,4,4).set_title('closing'),plt.imshow(cl_img, cmap = 'gray')
plt.subplot(2,4,5).set_title('opening'),plt.imshow(op_img, cmap = 'gray')
plt.subplot(2,4,6).set_title('top hat'),plt.imshow(th_img, cmap = 'gray')
plt.subplot(2,4,7).set_title('black hat'),plt.imshow(bh_img, cmap = 'gray')
plt.subplot(2,4,8).set_title('gradient'),plt.imshow(gd_img, cmap = 'gray');
```
