```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sobel_operator_image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# Sobel Filter

```python
sobelx = np.array([[-1,0,1],[-2,0,2], [-1,0,1]])
sobely =np.array([[1,2,1], [0,0,0], [-1,-2,-1]])
print('sobelx\n', sobelx)
print('\nsobely\n', sobely)
```

```python
sobelx = np.array([[-1,0,1],[-2,0,2], [-1,0,1]])
v_output = cv2.filter2D(img, ddepth = -1,  kernel =  sobelx)
plt.figure(figsize=(5,5))
plt.imshow(v_output)
plt.title(" Vertical Feature Extraction");
```

```python
h_output = cv2.filter2D(img, ddepth = -1,  kernel =  sobely)
plt.figure(figsize=(5,5))
plt.imshow(h_output)
plt.title(" Horizontal Feature Extraction");
```

```python
# combining both kernel to get both horizontal and vertical features
sobel = sobelx + sobely
c_output = cv2.filter2D(img, ddepth = -1,  kernel =  sobel)
plt.figure(figsize=(5,5))
plt.imshow(c_output)
plt.title(" Horizontal and Vertical Feature Extraction (adding kernel)");
```

```python
# another way is to add absolute value of vertical and horizontal output

combined = v_output + h_output
plt.figure(figsize=(5,5))
plt.imshow(combined)
plt.title(" Horizontal and Vertical Feature Extraction (adding extracted features)");
```

```python
# function sobelx , sobely and sobel
def sobelx(img):
    sobelx = np.array([[-1,0,1],[2,0,-2], [1,0,-1]])
    output = cv2.filter2D(img, ddepth = -1,  kernel =  sobelx)
    plt.figure(figsize=(5,5))
    plt.imshow(output, cmap = 'gray')
    plt.title("Vertical Feature Extraction")
    return 

def sobely(img):
    sobely =np.array([[-1,-2,-1], [0,0,0], [1,2,1]])
    output = cv2.filter2D(img, ddepth = -1,  kernel =  sobely)
    plt.figure(figsize=(5,5))
    plt.imshow(output, cmap = 'gray')
    plt.title("Horizontal Feature Extraction")
    return

def sobel(img):
    sobelx = np.array([[-1,0,1],[2,0,-2], [1,0,-1]])
    sobely =np.array([[-1,-2,-1], [0,0,0], [1,2,1]])
    sobel = sobelx +sobely
    output = cv2.filter2D(img, ddepth = -1,  kernel =  sobel)
    plt.figure(figsize=(5,5))
    plt.imshow(output, cmap = 'gray')
    plt.title(" Horizontal and Vertical Feature Extraction (adding kernel)")
    return 
```

# Laplace Filter  (6 of them)

```python
img3 = cv2.imread('images.png')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(5,5))
plt.imshow(img3, cmap = 'gray')
plt.title("Orginal Image");
```

```python
#laplace Kernels
k1 = np.array([[0,1,0],[1,-4,1], [0,1,0]]) # 2nd derivative
print('kernel 1 \n',k1)
print('\n')
nk1 =  np.negative(k1)
print('kernel 2 \n',nk1)
print('\n')

k2= np.array([[0,1,0],[1,-8,1], [0,1,0]]) # 4th derivative
print('kernel 3 (-8)\n',k2)
print('\n')

nk2 = np.negative(k2)
print('kernel 4 \n',nk2)
print('\n')

k3 = np.array([[0,1,0],[1,-5,1], [0,1,0]])  # 6th derivative
print('kernel 5 \n',k3)
print('\n')

nk3 = np.array([[-1,-1,-1],[-1,8,-1], [-1,8,-1]])
print('kernel 6 \n',nk3)
```

```python
#kernel 1
laplace_k1 =  cv2.filter2D(img3, ddepth = -1,  kernel =  k1)
plt.figure(figsize=(5,5))
plt.imshow(laplace_k1, cmap = 'gray' );
```

```python
#kernel 2
laplace_nk1 =  cv2.filter2D(img3, ddepth = -1,  kernel =  nk1)
plt.figure(figsize=(5,5))
plt.imshow(laplace_nk1 , cmap = 'gray');
```

```python
#kernel 3
laplace_k2 =  cv2.filter2D(img3, ddepth = -1,  kernel =  k2)
plt.figure(figsize=(5,5))
plt.imshow(laplace_k2, cmap = 'gray');
```

```python
#kernel 4
laplace_nk2 =  cv2.filter2D(img3, ddepth = -1,  kernel =  nk2)
plt.figure(figsize=(5,5))
plt.imshow(laplace_nk2,cmap = 'gray');
```

```python
#kernel 5
laplace_k3 =  cv2.filter2D(img3, ddepth = -1,  kernel =  k3)
plt.figure(figsize=(5,5))
plt.imshow(laplace_k3,cmap = 'gray' );
```

```python
#kernel 6
laplace_nk3=  cv2.filter2D(img3, ddepth = -1,  kernel =  nk3)
plt.imshow(laplace_nk3, cmap = 'gray');
```

```python
# plotting for comparison
plt.figure(figsize=(12,6))
    
plt.subplot(2,4,1).set_title('original'),plt.imshow(img3, cmap = 'gray')
plt.subplot(2,4,2).set_title('k1'),plt.imshow(laplace_k1, cmap = 'gray')
plt.subplot(2,4,3).set_title('k2'),plt.imshow(laplace_nk1, cmap = 'gray')
plt.subplot(2,4,4).set_title('k3'),plt.imshow(laplace_k2, cmap = 'gray')
plt.subplot(2,4,5).set_title('k4'),plt.imshow(laplace_nk2, cmap = 'gray')
#plt.subplot(2,4,6).set_title('k=5'),plt.imshow(th_img, cmap = 'gray')
plt.subplot(2,4,6).set_title('k5'),plt.imshow(laplace_k3, cmap = 'gray')
plt.subplot(2,4,7).set_title('k6'),plt.imshow(laplace_nk3, cmap = 'gray');
```

# Unsharp Masking

```python
img3 = cv2.imread('bean.jpg')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(4,4))
plt.imshow(img3)
plt.title('Orginal Image');
```

```python
blur = cv2.GaussianBlur(img3, (5, 5), 0)
med_blur = cv2.medianBlur(img3, 35)
sharp = img3+ (img3 - blur)
plt.imshow(sharp);
```

# Unsharp Masking Using PIL

```python
from PIL import Image, ImageFilter

img = Image.fromarray(img3.astype('uint8'))

sharp_img = img.filter(ImageFilter.UnsharpMask(radius=3, percent=100))

plt.figure(figsize=(12,8))
plt.subplot(121).set_title('original'), plt.imshow(img)
plt.subplot(122).set_title('Shapening filter'), plt.imshow(sharp_img);
```
